```python
# You are a financial analyst at a top investment firm. Your boss hands you a dataset 
# containing stock prices of 5 companies over a period of 30 days. Your task is to analyze 
# the data and extract meaningful insights! 
# �
# � Task: 
# 1. 
# 2. 
# 3. 
# 4. 
# 5. 
# Create a (30, 5) NumPy array with random stock prices between $100 and $500, 
# where: – – 
# Rows (0-29) represent days (30 days of stock prices). 
# Columns (0-4) represent 5 different companies. 
# Compute the average stock price for each company over the 30-day period. 
# Identify the highest stock price recorded in the dataset and the day and 
# company it belongs to. 
# Normalize the stock prices between 0 and 1 using the formula: 
# [ X_{} =  ] 
# Flag days where any stock drops below $200 as a “⏳ Risky Investment”. 

import numpy as np
dataset=np.random.randint(100,500,(30,5))
Average_stock_prices=np.mean(dataset,axis=0) 
avg=[]
for i in Average_stock_prices:
    avg.append(round(i,1))
print(f"Average stock prices:",avg)
maxi = np.max(dataset)
index = np.argwhere(dataset==maxi)
print(index)
print(f"Highest price recorded: {maxi} at Day {index[0][0]+1}, Company {index[0][1]}")
norm = np.linalg.norm(dataset)
matrix = dataset/norm
print("Normalized Prices : ")
print(matrix)
print("Risky Investments Days:")
for i in range(len(dataset)):
    mask = dataset[i] <200
    if len(dataset[i][mask]) !=0:
        print(f"Day {i+1} : {dataset[i][mask]}")
```

```python
# NASA is preparing a Mars mission ⏳ and needs to efficiently allocate oxygen, water, and 
# food for astronauts over 6 months. Your job as a space engineer is to analyze and optimize 
# resource consumption! 
# �
# � Task: 
# 1. 
# 2. 
# 3. 
# 4. 
# 5. 
# Create a (6, 3) NumPy array, where: – – 
# Rows (0-5) represent months. 
# Columns (0-2) represent oxygen, water, and food consumption (in tons). 
# Compute total resources required for the entire mission. 
# Identify which resource had the highest consumption in any given month and in 
# total. 
# Compute the standard deviation of each resource to measure consistency in 
# consumption. 
# Transpose the matrix to display the resource-wise breakdown by month.


import numpy as np
rand=np.random.randint(1,20,(6,3))
oxygen = np.sum(rand[:, 0])
water= np.sum(rand[:, 1])
food= np.sum(rand[:, 2])
print(f"Total resources needed(tons):Oxygen: {oxygen}, Water: {water}, Food: {food}")
print(f"Highest consumption in a month:")
max_values=np.max(rand,axis=0)
print(max_values)
max_months = np.argmax(rand, axis=0)+1 
print(max_months)
highest_resource_index = np.argmax([oxygen, water, food])
resource_names = ["Oxygen", "Water", "Food"]

print(f"Highest consumption in a month: {resource_names[np.argmax(max_values)]} ({max_values[np.argmax(max_values)]} tons in month {max_months[np.argmax(max_values)]})")
std_oxygen = round(np.std(rand[:, 0]), 1)
std_water = round(np.std(rand[:, 1]), 1)
std_food = round(np.std(rand[:, 2]),1)
print(f"Standard deviation of consumption: Oxygen: {std_oxygen}, Water: {std_water}, Food: {std_food}")
print("Transposed matrix:\n")
print(rand.T)
```
